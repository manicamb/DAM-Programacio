## Funcions proposades

Tot i que el projecte de LaOca és de lliure implementació, les següents funcions poden ser útils:

**clearScreen()**: neteja la pantalla

**assignarColor(fitxa)**: demana un nom a l’usuari, i el vàlida, per relacionar-lo amb la fitxa d’un color concret

**menuPrincipalDibuix()**: implementa la lògica de dibuix del menú principal, tenint en compte el nom dels jugadors inscrits i la disponibilitat de l'opció ‘Començar partida’

**menuPrincipal()**: gestiona la lògica del menú principal. 

És un bucle que mostra el dibuix del menú (amb menuPrincipalDibuix), demana una opció i comprova que les dades entrades són vàlides. 

També crida a les funcions ‘assignarColor’ o ‘jocJugar’

**taulellDibuixar()**: dibuixa el taulell de joc amb els jugadors a la posició (casella) on estàn

**promptEsborrar()**: esborra la llista que conté les linies d’informació que s’han d’escriure a sota del taulell (prompt)

**promptAfegir(frase)**: esborra el primer element de la llista que conté les línies d’informació del prompt, i afegeix una nova línia al final amb la frase que rep com a paràmetre

**promptDibuixar()**: dibuixa les línies de la llista que conté les frases del ‘prompt’ 

**dibuixar()**: esborra la pantalla, dibuixa el taulell i dibuixa el prompt

**jocIniciar()**: esborra les linies del prompt, reinicia les dades de la partida però manté els noms dels jugadors, defineix qui juga la primera tirada de la següent partida, esborra la llista del ranking de jugadors

**jocJugarDemanaTrampa(color)**: demana a quina casella es vol anar amb fent trampa

**jocJugarAccioCasella(tornJugador)**: 

- mira si el jugador ha arribat a la casella 53 i si és així si ha acabat la partida
- si el jugador està en una casella oca salta a la següent oca i torna a tirar
- si el jugador està en una casella daus salta a l’altre casella daus i torna a tirear
- si el jugador està en una casella pont, salta a l’altre casella pont
- si el jugador està en una casella hotel, defineix una espera d’1 torn
- si el jugador està en una casella pou, defineix una espera de 3 torns
- si el jugador està en una casella laberint, defineix una espera de 3 torns
- si el jugador està a la casella calabera, torna a la posició 0

**jocJugarEscriuRanking()**: un cop acabat el joc, escriu els textos del ranking al prompt

**jocJugarTorn(tornJugador)**: juga un torn del jugador, retorna dos valors si surt de la partida o si repetirà jugada (perquè cau a oca o daus)

- si el jugador està en espera (hotel, pou, laberint) mostra el missatge
- sino dibuixa el taulell, el prompt i demana quina acció es vol fer
- si es vol fer trampa, crida a jocJugarDemanaTrampa
- si es vol tirar el dau, tira el dau
- finalment mou el jugador a la nova posició (del dau o la trampa)
- s’executa el què toca per aquella casella (jocJugarAccioCasella)
- si la partida ha acabat
- escriu les dades del ranking al promp (jocJugarEscriuRanking())
- demana a l’usuari si vol tornar a jugar o vol sortir al menú
- retorna si l’usuari surt de la partida, o repeteix torn
