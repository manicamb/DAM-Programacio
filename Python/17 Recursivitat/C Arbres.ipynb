## Arbres

Les estructures d'arbre serveixen per representar jerarquies entre dades.

<center><img src="./assets/12b_00.png" height="250" style="max-height: 200px;" /></center>

Els arbres tenen una sèrie d'elements coneguts:

- Nodes: Cada element amb informació que forma part de l'arbre.
- Node arrel: Node que no té cap node pare.
- Node pare: Node que té un node fill.
- Node fill: Node que té un node pare.
- Node fulla: Node que no té cap node fill.
- Branca: Connecta un node pare amb un node fill.
- Camí: Conjunt de nodes que connecten un node amb un altre (A, B, E, K)

**Alçada** d'un arbre:

Nombre de nodes que hi ha des de la fulla més profunda fins a l'arrel.
L'exemple té alçada 4

**Pes** d'un arbre: 

Nombre de nodes que té l'arbre.
L'exemple té pes 15

**Ordre** d'un arbre:

Nombre màxim de nodes que té un node.
L'exemple té ordre 2, quan un arbre té ordre 2 també es diu que és un **arbre binari**.

## Arbres binaris

Els arbres binaris són estructures de dades que contenen un valor i dos fills, un fill esquerra i un fill dret.

```python
animals = {
    "node": "lleó",
    "esquerra": {
        "node": "girafa",
        "esquerra": {
            "node": "hipopòtam",
            "esquerra": None,
            "dreta": None
        },
        "dreta": None
    },
    "dreta": {
        "node": "zebra",
        "esquerra": {
            "node": "rinoceront",
            "esquerra": None,
            "dreta": None
        },
        "dreta": {
            "node": "elefant",
            "esquerra": None,
            "dreta": None
        }
    }
}
```
```text
L'estructura representa l'arbre següent:
              lleó
       /             \
   girafa           zebra
   /               /    \
 hipopòtam rinoceront elefant
```


**Preordre**  ['lleó', 'girafa', 'hipopòtam', 'zebra', 'rinoceront', 'elefant']

**Inordre** ['hipopòtam', 'girafa', 'lleó', 'rinoceront', 'zebra', 'elefant']

**Postordre** ['hipopòtam', 'girafa', 'rinoceront', 'elefant', 'zebra', 'lleó']
